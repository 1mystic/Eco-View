# EcoView — Knowledge Distillation Notebook
## ConvNeXt Teacher → MobileNetV3 Student → ONNX INT8

**Dataset**: TrashNet (6 waste classes) from Kaggle  
**Goal**: Train a fast, deployable student model via knowledge distillation from a high-accuracy teacher  
**Hardware**: Kaggle GPU T4 (free tier)  
**Output**: `student_model_quantized.onnx` — drop into `ecoview-ml-backend/` to enable real inference

---

### Pipeline Overview

```
TrashNet (6 classes, 2527 images)
  70% train / 15% valid / 15% test  [stratified]
          │
          ▼
  TEACHER: ConvNeXt Base (timm, ImageNet-22k pretrained)
           AdamW lr=3e-4, 15 epochs, CosineAnnealingLR
           Label smoothing 0.1, early stopping patience=3
          │  soft labels (T=4.0)
          ▼
  STUDENT: MobileNetV3 Large (timm)
           Loss = 0.7×CE + 0.3×KL×T²
           AdamW lr=1e-3, 10 epochs
          │
          ▼
  ONNX export (opset 17, dynamic batch)
          │
  onnxsim simplification
          │
  onnxruntime QInt8 dynamic quantization
          │
  student_model_quantized.onnx  (~6 MB)
```

### Note on Class Labels

TrashNet uses recycling categories. EcoView's production backend (`core/inference.py`) uses pollution categories.  
After training, update `LABELS` and `SEVERITY_MAP` in `ecoview-ml-backend/core/inference.py` to match the  
`class_mapping.json` produced by this notebook.

Alternatively, fine-tune on a pollution-specific dataset after initial distillation.

````python
# ── Global Configuration ──────────────────────────────────────────────────────
SEED = 42
IMG_SIZE = 224
BATCH_SIZE = 16  # T4 has 14.56 GB; ConvNeXt Base needs small batch in FP32 (AMP below lets student use 32)
NUM_CLASSES = 6
TEACHER_LR = 3e-4
STUDENT_LR = 1e-3
WEIGHT_DECAY = 1e-4
TEACHER_EPOCHS = 15
STUDENT_EPOCHS = 10
LABEL_SMOOTHING = 0.1
TEMPERATURE = 4.0
ALPHA = 0.7   # weight for hard (CE) loss
EARLY_STOP_PATIENCE = 3

# Kaggle dataset path — feyzazkefe/trashnet dataset
DATASET_ROOT = Path('/kaggle/input/datasets/feyzazkefe/trashnet/dataset-resized')

# Output directory for all artifacts
OUTPUT_DIR = Path('/kaggle/working')
OUTPUT_DIR.mkdir(exist_ok=True)

# Reproducibility
torch.manual_seed(SEED)
np.random.seed(SEED)

print(f'Config loaded. SEED={SEED}, BATCH={BATCH_SIZE}, T={TEMPERATURE}, α={ALPHA}')
````

## Section 2 — Dataset Validation

````python
# Discover all image files and their classes
CLASSES = sorted([d.name for d in DATASET_ROOT.iterdir() if d.is_dir()])
print(f'Classes found: {CLASSES}')
assert len(CLASSES) == NUM_CLASSES, f'Expected {NUM_CLASSES} classes, found {len(CLASSES)}'

# Build flat list of (path, label_idx)
all_paths, all_labels = [], []
class_to_idx = {c: i for i, c in enumerate(CLASSES)}

for cls in CLASSES:
    cls_dir = DATASET_ROOT / cls
    imgs = list(cls_dir.glob('*.jpg')) + list(cls_dir.glob('*.jpeg')) + list(cls_dir.glob('*.png'))
    all_paths.extend(imgs)
    all_labels.extend([class_to_idx[cls]] * len(imgs))
    print(f'  {cls:15s}: {len(imgs):4d} images')

print(f'\nTotal images: {len(all_paths)}')
print(f'Class mapping: {class_to_idx}')

# Save class mapping for deployment
idx_to_class = {v: k for k, v in class_to_idx.items()}
with open(OUTPUT_DIR / 'class_mapping.json', 'w') as f:
    json.dump({'class_to_idx': class_to_idx, 'idx_to_class': {str(k): v for k, v in idx_to_class.items()}}, f, indent=2)
print('Saved class_mapping.json')
````

````python
# Stratified 70/15/15 split
all_paths = [str(p) for p in all_paths]

train_paths, temp_paths, train_labels, temp_labels = train_test_split(
    all_paths, all_labels, test_size=0.30, stratify=all_labels, random_state=SEED
)
val_paths, test_paths, val_labels, test_labels = train_test_split(
    temp_paths, temp_labels, test_size=0.50, stratify=temp_labels, random_state=SEED
)

print(f'Train : {len(train_paths):5d} images ({len(train_paths)/len(all_paths)*100:.1f}%)')
print(f'Valid : {len(val_paths):5d} images ({len(val_paths)/len(all_paths)*100:.1f}%)')
print(f'Test  : {len(test_paths):5d} images ({len(test_paths)/len(all_paths)*100:.1f}%)')

# Verify stratification
for split_name, labels in [('Train', train_labels), ('Valid', val_labels), ('Test', test_labels)]:
    unique, counts = np.unique(labels, return_counts=True)
    fracs = counts / counts.sum()
    print(f'{split_name} class distribution: {dict(zip([CLASSES[u] for u in unique], [f"{f:.2f}" for f in fracs]))}')
````

## Section 4 — Data Pipeline & Augmentation

````python
# Albumentations transforms
train_transform = A.Compose([
    A.RandomResizedCrop(size=(IMG_SIZE, IMG_SIZE), scale=(0.7, 1.0)),
    A.HorizontalFlip(p=0.5),
    A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=15, p=0.5),
    A.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1, p=0.5),
    A.OneOf([
        A.GaussNoise(var_limit=(10.0, 50.0), p=1.0),
        A.ISONoise(p=1.0),
    ], p=0.2),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ToTensorV2(),
])

val_transform = A.Compose([
    A.Resize(height=IMG_SIZE, width=IMG_SIZE),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ToTensorV2(),
])

class TrashDataset(Dataset):
    def __init__(self, paths, labels, transform):
        self.paths = paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        img = np.array(Image.open(self.paths[idx]).convert('RGB'))
        aug = self.transform(image=img)
        return aug['image'], self.labels[idx]

train_ds = TrashDataset(train_paths, train_labels, train_transform)
val_ds   = TrashDataset(val_paths,   val_labels,   val_transform)
test_ds  = TrashDataset(test_paths,  test_labels,  val_transform)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,  num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_ds,   batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_ds,  batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

print(f'Train batches: {len(train_loader)}')
print(f'Valid batches: {len(val_loader)}')
print(f'Test  batches: {len(test_loader)}')
````

## Section 5 — Teacher Training (ConvNeXt Base)

````python
# Build teacher
teacher = timm.create_model(
    'convnext_base.fb_in22k_ft_in1k',
    pretrained=True,
    num_classes=NUM_CLASSES
).to(DEVICE)

param_count = sum(p.numel() for p in teacher.parameters()) / 1e6
print(f'Teacher: ConvNeXt Base | {param_count:.1f}M parameters')

# Optimizer + scheduler + loss
teacher_optimizer = AdamW(teacher.parameters(), lr=TEACHER_LR, weight_decay=WEIGHT_DECAY)
teacher_scheduler = CosineAnnealingLR(teacher_optimizer, T_max=TEACHER_EPOCHS)
teacher_criterion = nn.CrossEntropyLoss(label_smoothing=LABEL_SMOOTHING)
````

````python
def train_epoch(model, loader, optimizer, criterion, device, scaler=None):
    model.train()
    total_loss, correct, total = 0.0, 0, 0
    use_amp = scaler is not None and device.type == 'cuda'
    for imgs, labels in loader:
        imgs, labels = imgs.to(device), labels.to(device)
        optimizer.zero_grad()
        if use_amp:
            with torch.amp.autocast('cuda'):
                logits = model(imgs)
                loss = criterion(logits, labels)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            logits = model(imgs)
            loss = criterion(logits, labels)
            loss.backward()
            optimizer.step()
        total_loss += loss.item() * imgs.size(0)
        correct += (logits.argmax(1) == labels).sum().item()
        total += imgs.size(0)
    return total_loss / total, correct / total

@torch.no_grad()
def eval_epoch(model, loader, criterion, device):
    model.eval()
    total_loss, correct, total = 0.0, 0, 0
    for imgs, labels in loader:
        imgs, labels = imgs.to(device), labels.to(device)
        logits = model(imgs)
        loss = criterion(logits, labels)
        total_loss += loss.item() * imgs.size(0)
        correct += (logits.argmax(1) == labels).sum().item()
        total += imgs.size(0)
    return total_loss / total, correct / total
````

````python
# Train teacher with early stopping
teacher_history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
best_val_acc = 0.0
patience_counter = 0

# AMP GradScaler — halves activation memory, ~1.5x speedup on T4
teacher_scaler = torch.amp.GradScaler('cuda') if DEVICE.type == 'cuda' else None

print('Training TEACHER (ConvNeXt Base)...')
print('='*65)

for epoch in range(1, TEACHER_EPOCHS + 1):
    t0 = time.time()
    tr_loss, tr_acc = train_epoch(teacher, train_loader, teacher_optimizer, teacher_criterion, DEVICE, scaler=teacher_scaler)
    vl_loss, vl_acc = eval_epoch(teacher, val_loader, teacher_criterion, DEVICE)
    teacher_scheduler.step()

    teacher_history['train_loss'].append(tr_loss)
    teacher_history['val_loss'].append(vl_loss)
    teacher_history['train_acc'].append(tr_acc)
    teacher_history['val_acc'].append(vl_acc)

    elapsed = time.time() - t0
    print(f'Ep {epoch:2d}/{TEACHER_EPOCHS} | '
          f'tr_loss={tr_loss:.4f} tr_acc={tr_acc:.4f} | '
          f'vl_loss={vl_loss:.4f} vl_acc={vl_acc:.4f} | '
          f'{elapsed:.0f}s')

    # Save best
    if vl_acc > best_val_acc:
        best_val_acc = vl_acc
        torch.save(teacher.state_dict(), OUTPUT_DIR / 'teacher_best.pth')
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= EARLY_STOP_PATIENCE:
            print(f'Early stopping at epoch {epoch} (no improvement for {EARLY_STOP_PATIENCE} epochs)')
            break

print(f'\nBest teacher val accuracy: {best_val_acc:.4f}')
print('Saved teacher_best.pth')
````

## Section 7 — Knowledge Distillation (Student: MobileNetV3)

````python
# Build student
student = timm.create_model(
    'mobilenetv3_large_100',
    pretrained=True,
    num_classes=NUM_CLASSES
).to(DEVICE)

student_param_count = sum(p.numel() for p in student.parameters()) / 1e6
print(f'Student: MobileNetV3 Large | {student_param_count:.1f}M parameters')
print(f'Compression ratio: {param_count/student_param_count:.1f}x fewer params')

# Distillation loss
def distillation_loss(student_logits, teacher_logits, labels, T, alpha):
    """
    loss = alpha * CE(student, hard_labels)
         + (1 - alpha) * KL(student_soft, teacher_soft) * T^2
    """
    hard_loss = F.cross_entropy(student_logits, labels, label_smoothing=LABEL_SMOOTHING)
    soft_student = F.log_softmax(student_logits / T, dim=1)
    soft_teacher = F.softmax(teacher_logits / T, dim=1)
    soft_loss = F.kl_div(soft_student, soft_teacher, reduction='batchmean') * (T * T)
    return alpha * hard_loss + (1 - alpha) * soft_loss

student_optimizer = AdamW(student.parameters(), lr=STUDENT_LR, weight_decay=WEIGHT_DECAY)
student_scheduler = CosineAnnealingLR(student_optimizer, T_max=STUDENT_EPOCHS)
print(f'\nDistillation config: T={TEMPERATURE}, α={ALPHA}')
print(f'Loss = {ALPHA}×CE + {1-ALPHA}×KL×T²')
````

````python
def distill_epoch(student, teacher, loader, optimizer, T, alpha, device, scaler=None):
    student.train()
    teacher.eval()
    total_loss, correct, total = 0.0, 0, 0
    use_amp = scaler is not None and device.type == 'cuda'
    for imgs, labels in loader:
        imgs, labels = imgs.to(device), labels.to(device)
        with torch.no_grad():
            if use_amp:
                with torch.amp.autocast('cuda'):
                    t_logits = teacher(imgs)
            else:
                t_logits = teacher(imgs)
        optimizer.zero_grad()
        if use_amp:
            with torch.amp.autocast('cuda'):
                s_logits = student(imgs)
                loss = distillation_loss(s_logits, t_logits, labels, T, alpha)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            s_logits = student(imgs)
            loss = distillation_loss(s_logits, t_logits, labels, T, alpha)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        total_loss += loss.item() * imgs.size(0)
        correct += (s_logits.argmax(1) == labels).sum().item()
        total += imgs.size(0)
    return total_loss / total, correct / total

# AMP scaler for student distillation
student_scaler = torch.amp.GradScaler('cuda') if DEVICE.type == 'cuda' else None

# Train student via distillation
student_history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}
best_student_acc = 0.0

print('Training STUDENT via Knowledge Distillation...')
print('='*65)

for epoch in range(1, STUDENT_EPOCHS + 1):
    t0 = time.time()
    tr_loss, tr_acc = distill_epoch(student, teacher, train_loader, student_optimizer, TEMPERATURE, ALPHA, DEVICE, scaler=student_scaler)
    vl_loss, vl_acc = eval_epoch(student, val_loader, nn.CrossEntropyLoss(), DEVICE)
    student_scheduler.step()

    student_history['train_loss'].append(tr_loss)
    student_history['val_loss'].append(vl_loss)
    student_history['train_acc'].append(tr_acc)
    student_history['val_acc'].append(vl_acc)

    elapsed = time.time() - t0
    print(f'Ep {epoch:2d}/{STUDENT_EPOCHS} | '
          f'tr_loss={tr_loss:.4f} tr_acc={tr_acc:.4f} | '
          f'vl_loss={vl_loss:.4f} vl_acc={vl_acc:.4f} | '
          f'{elapsed:.0f}s')

    if vl_acc > best_student_acc:
        best_student_acc = vl_acc
        torch.save(student.state_dict(), OUTPUT_DIR / 'student_best.pth')

print(f'\nBest student val accuracy: {best_student_acc:.4f}')
print('Saved student_best.pth')
````

## Section 11 — INT8 Quantization

````python
from onnxruntime.quantization import quantize_dynamic, QuantType

quantized_path = OUTPUT_DIR / 'student_int8.onnx'
quantize_dynamic(
    model_input=str(simplified_path),
    model_output=str(quantized_path),
    weight_type=QuantType.QInt8,
    optimize_model=True,
)

q_size = quantized_path.stat().st_size / 1e6
print(f'Simplified FP32: {simp_size:.2f} MB')
print(f'Quantized INT8:  {q_size:.2f} MB  ({(1-q_size/simp_size)*100:.1f}% size reduction)')

# Validate quantized model accuracy
q_session = ort.InferenceSession(str(quantized_path), providers=['CPUExecutionProvider'])
q_input_name = q_session.get_inputs()[0].name

q_preds, q_true = [], []
for imgs, labels in test_loader:
    out = q_session.run(None, {q_input_name: imgs.numpy()})[0]
    q_preds.extend(out.argmax(1))
    q_true.extend(labels.numpy())

q_acc = np.mean(np.array(q_preds) == np.array(q_true))
s_acc = np.mean(np.array(all_preds_s) == np.array(all_true_s))
print(f'\nStudent FP32 accuracy:  {s_acc*100:.2f}%')
print(f'Student INT8 accuracy:  {q_acc*100:.2f}%  (Δ {(s_acc-q_acc)*100:+.2f}%)')
````

## Section 12 — Benchmarking

````python
def benchmark_pytorch(model, device, n_warmup=20, n_runs=100):
    img = torch.randn(1, 3, IMG_SIZE, IMG_SIZE, device=device)
    model.eval()
    with torch.no_grad():
        for _ in range(n_warmup):
            model(img)
    if device.type == 'cuda':
        torch.cuda.synchronize()
    t0 = time.perf_counter()
    with torch.no_grad():
        for _ in range(n_runs):
            model(img)
    if device.type == 'cuda':
        torch.cuda.synchronize()
    return (time.perf_counter() - t0) / n_runs * 1000  # ms

def benchmark_onnx(session, n_warmup=20, n_runs=100):
    img = np.random.randn(1, 3, IMG_SIZE, IMG_SIZE).astype(np.float32)
    iname = session.get_inputs()[0].name
    for _ in range(n_warmup):
        session.run(None, {iname: img})
    t0 = time.perf_counter()
    for _ in range(n_runs):
        session.run(None, {iname: img})
    return (time.perf_counter() - t0) / n_runs * 1000  # ms

cpu_device = torch.device('cpu')
teacher_cpu = timm.create_model('convnext_base.fb_in22k_ft_in1k', num_classes=NUM_CLASSES)
teacher_cpu.load_state_dict(torch.load(OUTPUT_DIR / 'teacher_best.pth', map_location='cpu', weights_only=True))
teacher_cpu.eval()

student_cpu = timm.create_model('mobilenetv3_large_100', num_classes=NUM_CLASSES)
student_cpu.load_state_dict(torch.load(OUTPUT_DIR / 'student_best.pth', map_location='cpu', weights_only=True))
student_cpu.eval()

fp32_sess = ort.InferenceSession(str(simplified_path), providers=['CPUExecutionProvider'])
int8_sess = ort.InferenceSession(str(quantized_path),  providers=['CPUExecutionProvider'])

print('Benchmarking (CPU, 100 runs after 20 warmups)...')
t_teacher = benchmark_pytorch(teacher_cpu, cpu_device)
t_student = benchmark_pytorch(student_cpu, cpu_device)
t_fp32    = benchmark_onnx(fp32_sess)
t_int8    = benchmark_onnx(int8_sess)

results = [
    ('Teacher (ConvNeXt, PyTorch)', param_count,        t_teacher,  best_val_acc,  (OUTPUT_DIR/'teacher_best.pth').stat().st_size/1e6),
    ('Student (MobileNetV3, PyTorch)', student_param_count, t_student, best_student_acc, (OUTPUT_DIR/'student_best.pth').stat().st_size/1e6),
    ('Student ONNX FP32',            student_param_count, t_fp32,   best_student_acc, simp_size),
    ('Student ONNX INT8',            student_param_count, t_int8,   q_acc,           q_size),
]

print(f'\n{"Model":<35} {"Params(M)":>10} {"Latency(ms)":>12} {"Val Acc":>8} {"Size(MB)":>9}')
print('-'*80)
for name, params, lat, acc, size in results:
    print(f'{name:<35} {params:>10.1f} {lat:>12.1f} {acc*100:>7.2f}% {size:>9.2f}')
````

After downloading `student_model_quantized.onnx` and `class_mapping.json` from Kaggle outputs:

### 2. Update class labels in `ecoview-ml-backend/core/inference.py`

Open `class_mapping.json` and note the class order (`idx_to_class`).
Update `LABELS` in `inference.py` to match — in index order:

```python
# Example for TrashNet:
LABELS = ['cardboard', 'glass', 'metal', 'paper', 'plastic', 'trash']

SEVERITY_MAP = {
    'cardboard': 'low',
    'glass':     'medium',
    'metal':     'medium',
    'paper':     'low',
    'plastic':   'high',
    'trash':     'high',
}
```

### 3. Redeploy HuggingFace Space

```bash
cd ecoview-ml-hf-space
git add student_model_quantized.onnx core/inference.py
git commit -m "add trained ONNX student model"
git push
```

The Space auto-rebuilds. Test with:
```bash
curl -X POST https://USERNAME-ecoview-ml.hf.space/inference/classify \
  -H 'Content-Type: application/json' \
  -d '{"image_url": "https://upload.wikimedia.org/wikipedia/commons/thumb/3/35/Single_random_blue_bin.jpg/640px-Single_random_blue_bin.jpg"}'
```

### 4. Enable ML classification in Report.jsx

Currently the frontend does not call `/inference/classify` because photos are stored as base64 (not HTTP URLs).  
To enable ML after this model upgrade, update the backend to accept base64 directly:

In `ecoview-ml-backend/main.py`:
```python
class ClassifyRequest(BaseModel):
    image_url: Optional[str] = None
    image_base64: Optional[str] = None  # add this field
```

In `ecoview-ml-backend/core/inference.py`, decode base64 before PIL:
```python
import base64
import io

if image_base64:
    # Strip data URL prefix if present
    if ',' in image_base64:
        image_base64 = image_base64.split(',', 1)[1]
    img_bytes = base64.b64decode(image_base64)
    img = Image.open(io.BytesIO(img_bytes)).convert('RGB')
```

Then re-enable the `classifyImage()` call in `frontend/src/pages/Report.jsx` after form submission.
